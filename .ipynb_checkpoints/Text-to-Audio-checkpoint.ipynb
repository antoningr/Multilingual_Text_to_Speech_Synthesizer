```python
import torch
from transformers import pipeline
import torchaudio
```

```python
# Charger le modèle Text-to-Speech
tts = pipeline("text-to-speech", model="facebook/mms-tts-eng")

# Texte à transformer en audio
text = "Hello, this is a test of text-to-speech synthesis using Hugging Face models."

# Générer l'audio
output = tts(text)

# Assurons-nous que l'audio est un tensor 2D (channels, time)
audio_tensor = torch.tensor(output['audio'])

# Vérification de la forme et suppression des dimensions inutiles si nécessaire
if audio_tensor.ndimension() == 3:
    audio_tensor = audio_tensor.squeeze(0)  # Retirer la dimension inutile (la première)

# Obtenir le taux d'échantillonnage
sample_rate = output['sampling_rate']

# Sauvegarde de l'audio au format WAV
torchaudio.save("output.wav", audio_tensor, sample_rate)

# Lecture de l'audio
import IPython.display as ipd
ipd.Audio("output.wav")
```
